```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_data.csv")
ride_data = pd.read_csv("Resources/ride_data.csv")
combined_data = pd.merge(city_data, ride_data, how='outer', on="city")

plot_data = combined_data
plot_data.head()
```

```python
urban_cities = plot_data[plot_data["type"]=="Urban"]
suburban_cities = plot_data[plot_data["type"]=="Suburban"]
rural_cities = plot_data[plot_data["type"]=="Rural"]

urban_ride_count = urban_cities["ride_id"].count()
suburban_ride_count = suburban_cities["ride_id"].count()
rural_ride_count = rural_cities["ride_id"].count()

slices = [urban_ride_count, rural_ride_count, suburban_ride_count]

labels = ["Urban", "Rural", "Suburban"]

colors = ["coral", "yellow", "aqua"]

explode = [.1, 0, 0]

plt.title("% of Total Rides by City Type")

plt.pie(slices, labels=labels, colors=colors, autopct="%1.1f%%", explode=explode, startangle=265)

plt.show()
```
