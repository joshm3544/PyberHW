```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_data.csv")
ride_data = pd.read_csv("Resources/ride_data.csv")
combined_data = pd.merge(city_data, ride_data, how='outer', on="city")

plot_data = combined_data
plot_data.head()
```

```python
urban_cities = plot_data[plot_data["type"]=="Urban"]
suburban_cities = plot_data[plot_data["type"]=="Suburban"]
rural_cities = plot_data[plot_data["type"]=="Rural"]

urban_fare_count = urban_cities["fare"].sum()
suburban_fare_count = suburban_cities["fare"].sum()
rural_fare_count = rural_cities["fare"].sum()

#total_fare_count = urban_fare_count + suburban_fare_count + rural_fare_count

#urban_fare_percent = (urban_fare_count / total_fare_count)
#suburban_fare_percent = (suburban_fare_count / total_fare_count)
#rural_fare_percent = (rural_fare_count / total_fare_count)

slices = [urban_fare_count, suburban_fare_count, rural_fare_count]

labels = ["Urban", "Suburban", "Rural"]

colors = ["red", "blue", "yellow"]

plt.pie(slices, labels=labels, colors=colors, autopct="%1.1f%%")

plt.show()
```
