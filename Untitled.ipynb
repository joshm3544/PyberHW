```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_data.csv")
ride_data = pd.read_csv("Resources/ride_data.csv")
combined_data = pd.merge(city_data, ride_data, how='outer', on="city")

plot_data = combined_data
plot_data.head()
```

```python
urban_cities = plot_data[plot_data["type"]=="Urban"]
suburban_cities = plot_data[plot_data["type"]=="Suburban"]
rural_cities = plot_data[plot_data["type"]=="Rural"]

urban_ride_count = urban_cities.groupby(["city"]).count()["ride_id"]
urban_average = urban_cities.groupby(["city"]).mean()["fare"]
urban_drivers = urban_cities.groupby(["city"]).mean()["driver_count"]

suburban_ride_count = suburban_cities.groupby(["city"]).count()["ride_id"]
suburban_average = suburban_cities.groupby(["city"]).mean()["fare"]
suburban_drivers = suburban_cities.groupby(["city"]).mean()["driver_count"]

rural_ride_count = rural_cities.groupby(["city"]).count()["ride_id"]
rural_average = rural_cities.groupby(["city"]).mean()["fare"]
rural_drivers = rural_cities.groupby(["city"]).mean()["driver_count"]
```

```python
plt.scatter(urban_ride_count, urban_average, s=10*urban_drivers, c="coral", edgecolor="b", linewidths=1, marker="o", alpha=0.8, label="Urban")
plt.scatter(suburban_ride_count, suburban_average, s=10*suburban_drivers, c="aqua", edgecolor="b", linewidths=1, marker="o", alpha=0.8, label="Suburban")
plt.scatter(rural_ride_count, rural_average, s=10*rural_drivers, c="orange", edgecolor="b",linewidths=1, marker="o", alpha=0.8, label="Rural")

plt.ylim(20, 40)

plt.xlim(0, 40)

plt.title("Pybar Ride Sharing Data (2016)")


plt.show()
```
