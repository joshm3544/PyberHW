```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_data.csv")
ride_data = pd.read_csv("Resources/ride_data.csv")
combined_data = pd.merge(city_data, ride_data, how='outer', on="city")

plot_data = combined_data
plot_data.head()
```

```python
drive_count = 100 * plot_data.groupby(["type"]).sum()["driver_count"] / plot_data["driver_count"].sum()

plt.pie(drive_count, labels=["Rural", "Suburban", "Urban"], colors=["yellow","aqua","coral"], explode=[0,0,.1], autopct="%1.1f%%", startangle=200)

plt.title("% of Total Drivers by City Type")

plt.show()
```
